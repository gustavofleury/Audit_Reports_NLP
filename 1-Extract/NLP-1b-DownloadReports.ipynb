```
print( len(dfLinks) )
dfLinksClean=dfLinks.drop_duplicates()
print( len(dfLinksClean) )
```

```
#Download Loop
path="/content/drive/My Drive/Reports/"
numberOfFiles = len(dfLinksClean)

for url in dfLinksClean["Links"]:  
  name_file = url.split("/",)[-1]
  local_file = path + name_file
  report_file = Path(local_file)
  print(report_file)
  numberOfFiles=numberOfFiles-1
  if report_file.is_file():
    print("File already exists: " + name_file + " Remaining files: " +  str(numberOfFiles))
  else:
    urllib.request.urlretrieve(url, local_file)
    print("Download: " + name_file + " Remaining files: " +  str(numberOfFiles) )
```
